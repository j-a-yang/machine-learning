```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

## Local Minimum (with "extra work" using quiver)
$f(x,y) = x^2 + y^2$

The minimum is at $(0, 0)$.

In the 2D contour plot, the min is indicated by a yellow triangle.

In the 3D contour plot, the min is indicated by a red triangle.

```
def f(x, y):
  return x**2 + y**2

x = np.arange(-10, 10)
y = np.arange(-10, 10)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.plot([0], [0], 'y^')
plt.show()
```

```
# Setting up the figure
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Plotting the graph using 3D contour plot and changing the viewing angle.
ax.contour3D(X, Y, Z, 50, cmap='binary')
plt.plot([0], [0], 'r^')
ax.view_init(50, 70)
plt.show()
```

```
# calculate partial derivatives
def partial_x(x):
  return 2.0 * x

def partial_y(y):
  return 2.0 * y

U = partial_x(X)
V = partial_y(Y)

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
plt.plot([0], [0], 'r^')
plt.show()
```

## Local Maximum (with "extra work" using quiver)

## Saddle Point (with "extra work" using quiver)
$f(x,y) = (x-5)^2 - y^2$

The minimum is at $(5, 0)$.

The point at which the gradient is zero is indicated by a yellow triangle on the 2D plot, and a red triangle on the 3D plot.  This point is not a minimum nor maximum.

```
def f(x, y):
  return (x-5)**2 - (y)**2

x = np.arange(-10, 10)
y = np.arange(-10, 10)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 100, cmap='RdGy')
plt.colorbar()
plt.plot([5], [0], 'y^')
plt.show()
```

```
# Setting up the figure
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Plotting the graph using 3D contour plot and changing the viewing angle.
ax.contour3D(X, Y, Z, 60, cmap='binary')
plt.plot([5], [0], 'r^')
ax.view_init(45, -75)
plt.show()
```

```
# calculate partial derivatives
def partial_x(x):
  return 2.0 * (x - 5)

def partial_y(y):
  return -2.0 * y

U = partial_x(X)
V = partial_y(Y)

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
plt.plot([5], [0], 'r^')
plt.show()
```
